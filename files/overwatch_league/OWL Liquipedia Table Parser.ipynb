# Reading Tables

The best way to get OWL standings is reading the Wikipedia HTML.

I downloaded each page by hand, did a little editing to pull out just the table, and then fixed a couple values by hand.

```python
COLUMNS_TO_SAVE = [
    "season",
    "region",
    "standing",
    "team",
    "match_record",
    "map_record",
    "map_differential",
]
```

```python
def keep_digits(s):
    output = ""
    for char in s:
        if char.isdigit():
            output += char

    return output
```

Have to `%s/colspan//g` on the file first and remove the header row.

```python
COLUMN_NAME_MAPPING = {
    "index": "standing",
    1: "division",
    2: "team",
    3: "match_record",
    4: "map_record",
    5: "map_differential",
}

standings_1_df = pd.read_html("./liquipedia_tables/s1.html")[0]

# The index has the right order, but the standings column has some cruft in it
standings_1_df = standings_1_df.reset_index()
standings_1_df["index"] += 1

standings_1_df = standings_1_df.dropna(how="all").rename(columns=COLUMN_NAME_MAPPING)

standings_1_df["season"] = 1
standings_1_df["region"] = "combined"
standings_1_df["standing"] = standings_1_df["standing"].astype("int32")

standings_1_df = standings_1_df[COLUMNS_TO_SAVE]

standings_1_df
```

```python
COLUMN_NAME_MAPPING = {
    "index": "standing",
    "Unnamed: 1": "division",
    "Unnamed: 2": "team",
    "Unnamed: 3": "match_record",
    "Unnamed: 4": "map_record",
    "Unnamed: 5": "map_differential",
}

standings_2_df = pd.read_html("./liquipedia_tables/s2.html")[0]

# The index has the right order, but the standings column has some cruft in it
standings_2_df = standings_2_df.reset_index()
standings_2_df["index"] += 1

standings_2_df = standings_2_df.dropna(how="all").rename(columns=COLUMN_NAME_MAPPING)

standings_2_df["season"] = 2
standings_2_df["region"] = "combined"

standings_2_df = standings_2_df[COLUMNS_TO_SAVE]

standings_2_df
```

```python
COLUMN_NAME_MAPPING = {
    "index": "standing",
    1: "team",
    2: "match_record",
    3: "map_record",
    4: "map_differential",
    "Unnamed: 1": "team",
    "Unnamed: 2": "match_record",
    "Unnamed: 3": "map_record",
    "Unnamed: 4": "map_differential",
}

na_standings_df, apac_standings_df = pd.read_html("./liquipedia_tables/s3.html")

na_standings_df["region"] = "north_america"
apac_standings_df["region"] = "apac"

na_standings_df = na_standings_df.dropna(how="all").rename(columns=COLUMN_NAME_MAPPING)
apac_standings_df = apac_standings_df.dropna(how="all").rename(
    columns=COLUMN_NAME_MAPPING
)

standings_3_df = pd.concat([na_standings_df, apac_standings_df])

# The index has the right order, but the standings column has some cruft in it
standings_3_df = standings_3_df.reset_index()
standings_3_df["index"] += 1

standings_3_df = standings_3_df.rename(columns=COLUMN_NAME_MAPPING)

standings_3_df["season"] = 3

standings_3_df = standings_3_df[COLUMNS_TO_SAVE]

standings_3_df
```

```python
COLUMN_NAME_MAPPING = {
    "index": "standing",
    1: "team",
    2: "match_record",
    3: "map_record",
    4: "map_differential",
    "Unnamed: 1": "team",
    "Unnamed: 2": "match_record",
    "Unnamed: 3": "map_record",
    "Unnamed: 4": "map_differential",
}

na_standings_df, apac_standings_df = pd.read_html("./liquipedia_tables/s4.html")

na_standings_df["region"] = "north_america"
apac_standings_df["region"] = "apac"

na_standings_df = na_standings_df.dropna(how="all").rename(columns=COLUMN_NAME_MAPPING)
apac_standings_df = apac_standings_df.dropna(how="all").rename(
    columns=COLUMN_NAME_MAPPING
)

standings_4_df = pd.concat([na_standings_df, apac_standings_df])

# The index has the right order, but the standings column has some cruft in it
standings_4_df = standings_4_df.reset_index()
standings_4_df["index"] += 1

standings_4_df = standings_4_df.rename(columns=COLUMN_NAME_MAPPING)

standings_4_df["season"] = 4

standings_4_df = standings_4_df[COLUMNS_TO_SAVE]

standings_4_df
```

```python
standings_df = pd.concat(
    [standings_1_df, standings_2_df, standings_3_df, standings_4_df]
)

standings_df
```

# Save as JSON

```python
standings_df.to_json("./owl_standings.json", orient="table")
```
