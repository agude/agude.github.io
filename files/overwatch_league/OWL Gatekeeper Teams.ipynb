# Data

```python
df = pd.read_csv("./match_map_stats.csv")

df["round_start_time"] = pd.to_datetime(df["round_start_time"])
df["round_end_time"] = pd.to_datetime(df["round_end_time"])

df.head()
```

```python
# Add the match loser
df["match_loser"] = np.where(
    df["team_one_name"] == df["match_winner"], df["team_two_name"], df["team_one_name"]
)

# Add year (there is a single match with NULL round_start_time,
# but all matches have non_NULL round_end_time so we use that.
df["year"] = df["round_end_time"].dt.year

# Convert year to season
df["season"] = df["year"] - 2017

df.head()
```

```python
game_df = (
    df.groupby(
        by=[
            "year",
            "season",
            "stage",
            "match_id",
            "game_number",
            "match_winner",
            "match_loser",
            "map_winner",
            "map_loser",
            "map_name",
            "winning_team_final_map_score",
            "losing_team_final_map_score",
        ],
        as_index=False,
    )
    .agg(
        game_start_time=("round_start_time", "min"),
        game_end_time=("round_end_time", "max"),
    )
    .sort_values(["game_start_time"])
)


game_df["match_winner_won_map"] = game_df["match_winner"] == game_df["map_winner"]
game_df["match_loser_won_map"] = game_df["match_loser"] == game_df["map_winner"]

game_df.head()
```

```python
match_df = game_df.groupby(
    by=[
        "year",
        "season",
        "stage",
        "match_id",
        "match_winner",
        "match_loser",
    ],
    as_index=False,
).agg(
    match_start_time=("game_start_time", "min"),
    match_end_time=("game_end_time", "max"),
    match_winner_maps_won=("match_winner_won_map", "sum"),
    match_loser_maps_won=("match_loser_won_map", "sum"),
)[  # Reorder output columns
    [
        "year",
        "season",
        "stage",
        "match_id",
        "match_start_time",
        "match_end_time",
        "match_winner",
        "match_loser",
        "match_winner_maps_won",
        "match_loser_maps_won",
    ]
]

match_df["match_winner_map_differential"] = (
    match_df["match_winner_maps_won"] - match_df["match_loser_maps_won"]
)
match_df["match_loser_map_differential"] = (
    match_df["match_loser_maps_won"] - match_df["match_winner_maps_won"]
)

match_df.head()
```

```python
match_df.to_json("./match_level_data.json", orient="table")
```

```python
standings_df = pd.read_json("./owl_standings.json", orient="table").drop(
    columns=["index"]
)

standings_df
```

```python
winner_standings_df = standings_df.rename(
    columns={
        "team": "match_winner",
        "standing": "winner_standing",
        "region": "winner_region",
    }
)[["season", "match_winner", "winner_region", "winner_standing"]]

loser_standings_df = standings_df.rename(
    columns={
        "team": "match_loser",
        "standing": "loser_standing",
        "region": "loser_region",
    }
)[["season", "match_loser", "loser_region", "loser_standing"]]
```

# Combine Data

```python
joined_df = match_df.merge(
    winner_standings_df, on=["season", "match_winner"], how="inner"
).merge(loser_standings_df, on=["season", "match_loser"], how="inner")

joined_df = joined_df.sort_values(["match_start_time"]).reset_index(drop=True)

joined_df["winner_standing"] = pd.to_numeric(joined_df["winner_standing"])
joined_df["loser_standing"] = pd.to_numeric(joined_df["loser_standing"])

joined_df["higher_ranked_team_won"] = (
    joined_df["winner_standing"] < joined_df["loser_standing"]
)

joined_df.head()
```

```python
def compute_gatekeeper_score(team, season, df):
    # Get just the rows we care about
    is_right_seaon = df["season"] == season
    is_right_team = (df["match_winner"] == team) | (df["match_loser"] == team)
    is_same_region = df["winner_region"] == df["loser_region"]

    pruned_df = df[is_right_seaon & is_right_team & is_same_region]

    # Compute fractions
    our_team_won = pruned_df["match_winner"] == team
    our_team_lost = pruned_df["match_loser"] == team
    higher_rank_won = pruned_df["higher_ranked_team_won"] == True
    higher_rank_lost = pruned_df["higher_ranked_team_won"] == False

    n_won_against_worse = len(pruned_df[our_team_won & higher_rank_won])
    n_lost_against_worse = len(pruned_df[our_team_lost & higher_rank_lost])

    n_won_against_better = len(pruned_df[our_team_won & higher_rank_lost])
    n_lost_against_better = len(pruned_df[our_team_lost & higher_rank_won])

    # Sometimes we have divide by zero, which should be nan
    try:
        win_rate_against_worse = n_won_against_worse / (
            n_won_against_worse + n_lost_against_worse
        )
    except ZeroDivisionError:
        win_rate_against_worse = float("nan")

    try:
        win_rate_against_better = n_won_against_better / (
            n_won_against_better + n_lost_against_better
        )
    except ZeroDivisionError:
        win_rate_against_better = float("nan")

    # Compute differentials
    map_differential_against_worse = (
        pruned_df[our_team_won & higher_rank_won]["match_winner_map_differential"].sum()
        + pruned_df[our_team_lost & higher_rank_lost][
            "match_loser_map_differential"
        ].sum()
    )
    map_differential_against_better = (
        pruned_df[our_team_won & higher_rank_lost][
            "match_winner_map_differential"
        ].sum()
        + pruned_df[our_team_lost & higher_rank_won][
            "match_loser_map_differential"
        ].sum()
    )

    gatekeeper_score = win_rate_against_worse - win_rate_against_better
    gatekeeper_differential = (
        map_differential_against_worse - map_differential_against_better
    )

    return (
        win_rate_against_worse,
        win_rate_against_better,
        gatekeeper_score,
        map_differential_against_worse,
        map_differential_against_better,
        gatekeeper_differential,
    )
```

```python
compute_gatekeeper_score("Los Angeles Valiant", 1, joined_df)
```

```python
compute_gatekeeper_score("New York Excelsior", 1, joined_df)
```

```python
compute_gatekeeper_score("Atlanta Reign", 3, joined_df)
```

```python
pruned_df = standings_df[["team", "season", "standing"]]

standings_df["win_rate_against_worse"] = None
standings_df["win_rate_against_better"] = None
standings_df["gatekeeper_score"] = None
standings_df["map_differential_against_worse"] = None
standings_df["map_differential_against_better"] = None
standings_df["gatekeeper_differential"] = None


for season, team in zip(pruned_df["season"], pruned_df["team"]):
    (
        win_rate_against_worse,
        win_rate_against_better,
        gatekeeper_score,
        map_differential_against_worse,
        map_differential_against_better,
        gatekeeper_differential,
    ) = compute_gatekeeper_score(team, season, joined_df)
    right_season_and_team = (standings_df["season"] == season) & (
        standings_df["team"] == team
    )

    try:
        rounded_score = int(round(gatekeeper_score, 2) * 100)
    except ValueError:
        rounded_score = float("nan")

    standings_df.loc[right_season_and_team, "win_rate_against_worse"] = round(
        win_rate_against_worse, 3
    )
    standings_df.loc[right_season_and_team, "win_rate_against_better"] = round(
        win_rate_against_better, 3
    )
    standings_df.loc[right_season_and_team, "gatekeeper_score"] = rounded_score
    standings_df.loc[
        right_season_and_team, "map_differential_against_worse"
    ] = map_differential_against_worse
    standings_df.loc[
        right_season_and_team, "map_differential_against_better"
    ] = map_differential_against_better
    standings_df.loc[
        right_season_and_team, "gatekeeper_differential"
    ] = gatekeeper_differential
```

```python
standings_df.sort_values(["gatekeeper_score"], ascending=False).reset_index()
```

```python
standings_df.to_json("./combined_standings_data.json", orient="table")
```
