```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sqlite3
import seaborn as sns
from matplotlib import cm
```

```python
# Set plotting style
plt.style.use('seaborn-white')

LINE_COLOR = "green"

%matplotlib inline
```

```python
START_YEAR = 2001
LAST_YEAR = 2017
```

```python
# Date conversion function
def date_columns(query):
    """If a date column is included in the query, parse it as a date in the
    dataframe."""
    dates = []
    fields = ["Collision_Date", "Process_Date"]
    if '*' in query:
        dates = fields
    else:
        for date in fields:
            if date in query:
                dates.append(date)

        if not dates:
            dates = None

    return dates
```

```python
def run_query(query, sql_file="./20180925_switrs.sqlite3"):
    """ Read sqlite query results into a pandas DataFrame. """
    with sqlite3.connect(sql_file) as con:
        # Construct a Dataframe from the results
        df = pd.read_sql_query(
            query,
            con,
            parse_dates = date_columns(query),
        )

        return df
```

# Crashes by Week

Let's start by looking at the number of incidents per month:

```python
query = (
    "SELECT Collision_Date "
    "FROM Collision "
    "WHERE Collision_Date IS NOT NULL "
    "AND Collision_Date <= '{}-12-31' "
    "AND Bicycle_Collision == 1 ".format(LAST_YEAR)
)

df = run_query(query)
```

```python
DATE_COL = "Collision_Date"
CRASH_COL = "Crashes"

df["DOY"]     = df[DATE_COL].dt.dayofyear
df["DOW"]     = df[DATE_COL].dt.dayofweek
df["Day"]     = df[DATE_COL].dt.day
df["Week"]    = df[DATE_COL].dt.week
df["Month"]   = df[DATE_COL].dt.month
df["Year"]    = df[DATE_COL].dt.year
df[CRASH_COL] = 1
```

```python
# Convert to a timeseries
ts = df[[DATE_COL]]
ts.index= ts[DATE_COL]
ax = ts.resample('W-MON').count()[DATE_COL].plot(
    kind="line",
    title='Crashes in California',
    figsize=(12,6),
    linewidth=1,
    color=LINE_COLOR,
)

# Set Y range and grid
ax.grid()
ax.set_xlim((pd.Timestamp("2001-01-01"), pd.Timestamp("2017-12-31")))
ax.set_ylim((75, 420))

# Set axis labels
plt.title('Bicycle Crashes in California', y=1.03, size=28)
FONTSIZE="xx-large"
plt.xlabel("")
plt.ylabel("Crashes / Week", fontsize=FONTSIZE)

# Set the year between the tick marks using minor ticks
# Pandas uses a obscure Axis labeling system, so to put the ticks between the
# major ticks, we interpolate it using np.interp().
y = ax.get_xlim()
x = (START_YEAR - 0.5, LAST_YEAR + 0.5)

minor_x = np.interp(np.arange(START_YEAR, LAST_YEAR + 2, 2), x, y)
minor_label = np.arange(START_YEAR, LAST_YEAR + 2, 2)

ax.set_xticks(minor_x, minor=True)
ax.set_xticklabels(minor_label, minor=True, size=14)
ax.set_xticks([])  # Unset major ticks

# Add shaded bands for every other year
for year in range(START_YEAR + 1, LAST_YEAR + 3, 2):
    ax.axvspan(pd.to_datetime(str(year)), pd.to_datetime(str(year+1)), color="black", alpha=0.05)

for ext in ("png", "svg"):
    plt.savefig("/tmp/bicycle_accidents_per_week_in_california.{ext}".format(ext=ext), bbox_inches="tight")
```

# Grouped by Day of the Year

```python
from datetime import datetime

def annotate_year(df, ax, month, day, text, xytext, adjust=(0, 0), arrowstyle="->"):
    """ Draw an annotation on the Day of Year plot. """
    # Use 2016 because it is a leapyear, and we plot Feb 29
    doy = datetime(year=2016, month=month, day=day).timetuple().tm_yday
    y_pos = df[CRASH_COL][month][day]

    ax.annotate(
        text,
        (doy+adjust[0], y_pos+adjust[1]),
        xytext=xytext,
        textcoords='offset points',
        arrowprops=dict(
            arrowstyle=arrowstyle,
            connectionstyle="arc3",
        ),
        size=16,
    )
```

```python
# Get the start locations of each month
def month_starts(df):
    """ Get the start and midpoints of each month. """
    # Month starts
    majors = []
    for x, (month, day) in enumerate(df.index):
        if day == 1:
            majors.append(x)
        if month == 12 and day == 31:
            majors.append(x)

    # Midpoints
    minors = []
    for i in range(len(majors)-1):
        end = majors[i+1]
        start = majors[i]
        x = start + (end-start)/2.
        minors.append(x)

    return (majors, minors)
```

```python
# Calculate crashes per day
NUMBER_OF_YEARS = (LAST_YEAR - START_YEAR) + 1
LEAP_YEAR_ADJUSTMENT = NUMBER_OF_YEARS/4.

mean_crashes = df.groupby(["Month", "Day"]).count()
mean_crashes[CRASH_COL] /= NUMBER_OF_YEARS  # Average instead of sum of years
mean_crashes[CRASH_COL][2][29] = mean_crashes[CRASH_COL][2][29] * LEAP_YEAR_ADJUSTMENT
```

```python
ax = mean_crashes[CRASH_COL].plot(
    kind="line",
    linewidth=1.8,
    figsize=(12,6),
    color=LINE_COLOR,
)

# Set Y Range and grid
ax.set_xlim([-2, 367])  # Avoid hiding the start/end points
ax.set_ylim([5, 55])  # Avoid hiding the start/end points
ax.grid()

# Set axis labels
plt.title('Mean Bicycle Crashes by Date ({}–{})'.format(START_YEAR, LAST_YEAR), y=1.03, size=28)
FONTSIZE="xx-large"
plt.xlabel('')
plt.ylabel("Crashes", fontsize=FONTSIZE)

# Fix the X tick labels
(major_x, minor_x) = month_starts(mean_crashes)
labels = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
ax.set_xticks(major_x[:-1])
ax.set_xticklabels([])
ax.set_xticks(minor_x, minor=True)
ax.set_xticklabels(labels, minor=True, size=11)

# Shade every other month
for i in range(0, len(major_x)-1, 2):
    start = major_x[i+1]
    end = major_x[i+2]
    ax.axvspan(start, end, color="black", alpha=0.03)

# Federal Holidays
annotate_year(mean_crashes, ax, 1, 1, "New Years", (5, -40), (-2, 0))
annotate_year(mean_crashes, ax, 1, 18, "MLK", (-17, 25))
annotate_year(mean_crashes, ax, 2, 18, "Washington's Birthday", (-79.5, -27), (0, -1.5), arrowstyle="-[")
annotate_year(mean_crashes, ax, 5, 28, "Memorial Day", (-49.5, -30), (0, -1), arrowstyle="-[")
annotate_year(mean_crashes, ax, 7, 4, "4th of July", (-20, +45))
annotate_year(mean_crashes, ax, 9, 4, "Labor Day", (-37, -35), (0, -2), arrowstyle="-[")
annotate_year(mean_crashes, ax, 10, 11, "Columbus Day", (-52, +30), (0, +7), arrowstyle="-[")
annotate_year(mean_crashes, ax, 11, 11, "Veterans Day", (-80, -25))
annotate_year(mean_crashes, ax, 11, 25, "Thanksgiving", (-47.5, -32), (0, -5), arrowstyle="-[")
annotate_year(mean_crashes, ax, 12, 25, "Christmas", (-90, -0))

# Other Holidays
annotate_year(mean_crashes, ax, 10, 31, "Halloween", (+5, +35))
annotate_year(mean_crashes, ax, 2, 14, "Valentine's Day", (-55, 50))
annotate_year(mean_crashes, ax, 3, 17, "St. Patrick's Day", (-40, 60))

for ext in ("png", "svg"):
    plt.savefig("/tmp/mean_bicycle_accidents_by_date.{ext}".format(ext=ext), bbox_inches="tight")
```

# Weekends

```python
# Count accidents per day, and then reset the index so we can pivot on the index
DOW_COL = "DOW"

df_dow = df[[DATE_COL, DOW_COL, "Crashes"]].groupby([DATE_COL, DOW_COL]).count()
df_dow = df_dow.reset_index()
```

```python
# Set plotting style
plt.style.use('seaborn-white')

# Set plot size
fig = plt.gcf()
fig.set_size_inches(12, 6)

colors = cm.viridis(np.linspace(0.4, 1, 7))
ax = sns.violinplot(
    x="DOW",
    y="Crashes",
    data=df_dow,
    linewidth=2,
    cut=0,
    inner="quartile",
    palette=colors,
)

# Set plot title
plt.title('Bicycle Crashes', y=1.03, size=28)

# Set Y and X range
#plt.gca().set_ylim([0, 3000])
plt.ylabel("Crashes", fontsize=FONTSIZE)
plt.xlabel("")

# Set X tick labels
ax.set_xticklabels(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"], size=14)

for ext in ("png", "svg"):
    plt.savefig("/tmp/bicycle_accidents_by_day_of_the_week.{ext}".format(ext=ext), bbox_inches="tight")
```
